## 1.核心本质：地址与指向

链表的储存类型(分散的)决定了我们并不更改其内部储存的数据，而是创造节点和更改指针指向。

- 我们手里的值(val)
- 节点的指向(next)

判断两个节点是否相等的逻辑与数组完全不同，用的是两个节点的地址，而不是值的大小

---

## 2. 两个工具
**工具A:dummy node**

- **用途**：保证头节点不会错乱，只要涉及到头节点的**位置变动或者删除**，通通先建立一个`dummy`
- **逻辑**：dummy.next = head

有了dummy我们可以不用写`if head is None`，最后统一返回`return dummy.next`

**工具B:三剑客指针(prev,cur,nxt)**
- 用于翻转或者两两交换
- 先用`nxt`链接要断开的节点，防止丢失后续的链，再修改`curr.next`的指向，最后全员平移

## 3.三种技巧(双指针)


#### **① 快慢指针 (追及问题)**

* **代表题**：141/142 环形链表。
* **秘籍**：`fast` 走两步，`slow` 走一步。相遇即有环。
* **进阶**：找入口时，一个从起点走，一个从相遇点走，相遇处即入口。

#### **② 距离指针 (间距问题)**

* **代表题**：19. 删除倒数第 N 个节点。
* **秘籍**：让 `fast` 先跑  步，保持恒定距离。当 `fast` 撞线（`None`）时，`slow` 刚好站在被删节点的前面。

#### **③ 换位对齐 (长度差问题)**

* **代表题**：160. 相交链表。
* **秘籍**：我走过你走过的路。`pA` 走完 A 接 B，`pB` 走完 B 接 A。
* **本质**：用路程的对称性 () 抵消长度差。
