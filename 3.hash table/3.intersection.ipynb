# 349.两个数组的交集

[力扣题目链接](https://leetcode.cn/problems/intersection-of-two-arrays/)

题意：给定两个数组，编写一个函数来计算它们的交集。



输出的每一个元素一定是唯一的，可以不考虑输出的顺序。

**思路**

该题无法用数组，因为没有限制数值的大小，使用数组不划算

我们可以使用集合(set)，只关心在不在，且天生自带"去重"功能。

这道题我们只关心一个数是否在两个数组里都出现过。不在乎它出现了几次，也不关心他的下标。

1. 把`nums1`变成一个集合`set1`
2. 遍历`nums2`中的每个元素
3. 判断这个元素是否在`set1`
4. 如果在就存入新的结果集合`res`中
5. 最后将`res`转成数组返回


````python
class Solution(object):
    def intersection(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        set1 = set(nums1)
        res = set()
        for i in nums2:
            if i in set1:
                res.add(i)

        return list(res)

#or
class Solution(object):
    def intersection(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """

        return list(set(nums1) & set(nums2))
        
````
