```python
class Solution(object):
    def swapPairs(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        dummy = ListNode(0)
        dummy.next = head
        temp = dummy #
        while temp.next and temp.next.next:#
            #node
            node1 = temp.next
            node2 = temp.next.next
            #templist
            temp.next = node2
            #node1
            node1.next =node2.next
            #node2
            node2.next = node1
            
            #tempnode1node1node2
            temp = node1

        return dummy.next

        
            
```
