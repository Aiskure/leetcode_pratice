# 151.翻转字符串里的单词

[力扣题目链接](https://leetcode.cn/problems/reverse-words-in-a-string/)

给定一个字符串，逐个翻转字符串中的每个单词。
<br/>
<br/>
<br/>
示例 1：

输入: "the sky is blue"

输出: "blue is sky the"
<br/>
<br/>
<br/>
示例 2：

输入: "  hello world!  "

输出: "world! hello"

解释: 输入字符串可以在前面或者后面包含多余的空格，但是反转后的字符不能包括。
<br/>
<br/>
<br/>
示例 3：

输入: "a good   example"

输出: "example good a"

解释: 如果两个单词间有多余的空格，将反转后单词间的空格减少到只含一个。

---

可以直接用`split()`，但是体现不出思路。

我们尝试用双指针

<br/>

**核心挑战**：多余空格
- 字符串开头和结尾可能有空格
- 单词之间可能有多个连续空格
- **目标**：翻转之后，但此间只能有一个空格，且首位无空格

<br/>

1. 移除多余的空格:`" hello world" `变成`"hello world"`
2. 翻转整个字符串： 变成`"dlrow olleh"`
3. 反转每个单词：变成`"world hello"`

<br/>

移除空格我们可以用快慢指针

- 快指针`fast`：扫描整个原始字符串
- 慢指针`slow`：指向我们需要保留的字符串

逻辑：
1. 跳过所有空格
2. 遇到单词时：
    - 如果`slow`不在开头，说明前面已经有一个单词了，我们需要先手动加一个空格`res[slow] = " "`
    - 然后把整个单词从`fast`搬运到`slow`的位置

    




```python
#直接用spilt
class Solution(object):
    def reverseWords(self, s):     
        return " ".join(s.split()[::-1])

#用快慢指针
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
#----------------------------------1.删除空格----------------------------
        #我们用快指针跳过空格，慢指针来填充新的内容，最后再用切片来保留慢指针
        chars = list(s)
        fast,slow = 0,0

        while fast < len(chars):
            if chars[fast] != " ": #fast直接跳过空格，只会有单词
                if slow != 0 : #如果我们要填充的单词不是首个单词
                    chars[slow] = ' ' #我们要填充一个空格
                    slow+=1
                
                while fast < len(chars) and chars[fast] != ' ': #搬运单词
                    chars[slow] = chars[fast]
                    slow +=1
                    fast +=1
                    #当fast移动到空格处，while终止，进入外部循环
            # 进入外部循环，此时是空格，也不用进入if，我们直接跳过空格
            fast +=1
        #切片
        chars = chars[:slow]
#----------------------------------2.翻转字符串--------------------------------
        self.reverse_range(chars, 0, len(chars) - 1)

#----------------------------------3.翻转每个单词--------------------------------
        start = 0
        for end in range(len(chars) + 1):
            #遇到空格或者末尾，表示这个单词结束了
            if end == len(chars) or chars[end] ==' ':
                self.reverse_range(chars,start,end-1)
                start = end + 1
        return "".join(chars)
        
    def reverse_range(self, l, left, right):
        while left < right:
            l[left], l[right] = l[right], l[left]
            left += 1
            right -= 1


        
    
```
