# 242.有效的字母异位词
[力扣题目链接](https://leetcode.cn/problems/valid-anagram/)

给定两个字符串 s 和 t ，编写一个函数来判断 t 是否是 s 的字母异位词。

示例 1: 输入: s = "anagram", t = "nagaram" 输出: true

示例 2: 输入: s = "rat", t = "car" 输出: false

说明: 你可以假设字符串只包含小写字母。

--- 

思路：

建立一个26位的数组，分别代表a到z

1. 记录s字符串，每遇到一个字母就在对应的抽屉里`+1`
2. 消耗t字符串，每遇到一个字母就在对应的抽屉里`-1`
3. 检查26个抽屉是不是全部变成0

tips：

我们用ord来将字母转化为数字
- ord(char) - ord('a')，如果char是"a"：97-97 = 0，数组排在第0位，以此类推

检查s和t长度是否一致

```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        #如果长度不一致肯定不是异位符
        if len(s) !=len(t):
            return False

        record = [0]*26

        for char in s:
            record[ord(char)-ord('a')] +=1

        for char in t:
            record[ord(char)-ord('a')] -=1

        for i in range(26):
            if record[i] !=0:
                return False
        return True

```
