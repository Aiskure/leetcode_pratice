# 150. 逆波兰表达式求值 (Evaluate Reverse Polish Notation)

[力扣题目链接](https://leetcode.cn/problems/evaluate-reverse-polish-notation/description/)

### 1. 题目描述
给你一个字符串数组 `tokens` ，表示一个根据 **逆波兰表示法** 表示的算术表达式。
请你计算该表达式。返回一个表示表达式值的整数。

**有效的算符为：** `+`、`-`、`*`、`/` 。
每个操作数可以是整数，也可以是另一个表达式。

**注意：**
* 两个整数之间的除法总是 **向零截断** （例如 `13 // 5 = 2`, `-13 // 5 = -2`）。
* 输入的逆波兰表达式总是有效的。这意味着表达式总会得出有效数值且不存在除数为 0 的情况。

---

### 2. 示例
* **示例 1：**
    * **输入：** `tokens = ["2", "1", "+", "3", "*"]`
    * **输出：** `9`
    * **解释：** 该算式转化为常见的中缀表达式为：`((2 + 1) * 3) = 9`

* **示例 2：**
    * **输入：** `tokens = ["4", "13", "5", "/", "+"]`
    * **输出：** `6`
    * **解释：** 该算式转化为：`(4 + (13 / 5)) = 6`

* **示例 3：**
    * **输入：** `tokens = ["10", "6", "9", "3", "+", "-11", "*", "/", "*", "17", "+", "5", "+"]`
    * **输出：** `22`

### 3. 逆波兰表达式（后缀表达式）简介
我们平时写的 `a + b` 叫 **中缀表达式**。而逆波兰表达式（RPN）是将运算符写在操作数后面，例如 `a b +`。
* **优点：** 不需要括号来决定优先级，计算机扫描时非常方便。
* **核心规律：** 每当遇到一个**运算符**，它就作用于它**前面最近的两个数字**。

```python
from operator import add, sub, mul

def div(x, y):
    # 使用整数除法的向零取整方式
    return int(x / y) if x * y > 0 else -(abs(x) // abs(y))

class Solution(object):
    op_map = {'+': add, '-': sub, '*': mul, '/': div}
    def evalRPN(self, tokens):
        
        """
        :type tokens: List[str]
        :rtype: int
        """
        
        stack = []

        for char in tokens:
            if char not in ["+","-","*","/"]:
                stack.append(int(char))
            else:
                num1 = stack.pop()
                num2 = stack.pop()
                stack.append(self.op_map[char](num2,num1))
        return stack.pop()

        
```
