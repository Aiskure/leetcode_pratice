### 1.stack--先进后出(LIFO)
栈stack是一种线性数据结构，它遵循特定的操作顺序。这个顺序是后进先出（LIFO）。后进先出意味着最后插入的元素最先被移出，而先进后出意味着最先插入的元素最后被移出。<br/>
它的行为就像一堆盘子，最后放上去的盘子最先被拿走。<br/>
可以这样理解：
- (push)将元素压入栈就像在盘子堆的最上面放一个新盘子。
- (pop)弹出元素就是从栈顶移除一个盘子。

- (top)查看最上面的元素

## 2. 队列 (Queue) —— 先进先出 ()

队列就像排队买票，讲究的是**先来后到**。

### 核心特性：

* **进出规则**：**FIFO** (First In, First Out)。最先排队的，最先买票走人。
* **主要操作**：
1. **Enqueue (入队/Push)**：从**队尾**加入。
2. **Dequeue (出队/Pop)**：从**队头**离开。
3. **Front (查看队头)**：查看下一个要离开的人。



* **物理模型**：

### 常见变体：

* **双端队列 (Deque)**：头尾都能进，头尾都能出（这是个“全能选手”，后面我们会经常用到）。

---

## 3. 栈与队列的对比

为了让你直观感受，我们可以看这个对比表：

| 特性 | 栈 (Stack) | 队列 (Queue) |
| --- | --- | --- |
| **规则** | 后进先出 (LIFO) | 先进先出 (FIFO) |
| **操作点** | 同一个口（栈顶） | 两个口（队头出，队尾进） |
| **现实类比** | 堆叠的盘子、撤销功能 | 银行排队、打印机任务 |
| **常用场景** | 递归、括号匹配、深度优先搜索 (DFS) | 任务调度、层级遍历、广度优先搜索 (BFS) |

## 4. 引导思考

在 Python 中，我们通常使用 `list` 来模拟栈：

* `stack.append(x)` 是入栈。
* `stack.pop()` 是出栈。

**提问引导：**
如果我们要用 Python 的 `list` 来模拟**队列**：

1. `list.append(x)` 可以作为入队（往队尾加）。
2. 那么出队（从队头拿）应该用什么操作？
3. 提示：回想一下数组专题，**从数组头部删除元素**，它的时间复杂度是多少？这会是一个高效的队列吗？




list.pop(0)可以完成queue的出列，但是每次出列需要把数组往前移动`n-1`，时间复杂度是$O(n)$，数据很大时会很慢。<br/>

在python中我们用`collections.deque`，他的底层不是简单的连续数组，而是类似于链表的结构。

- 入队：`append` - $O(1)$
- 出列：`popleft()` -$O(1)$(直接掐掉头节点)

```python
from collections import deque

d = deque()         #空队列

d =deque([1,2,3],maxlen = 5) #限制最大长度5

d = deque([1,2,3])  #从可迭代对象初始化

```
