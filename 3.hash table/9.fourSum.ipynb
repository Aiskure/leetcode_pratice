# 18.四数之和

[力扣题目链接](https://leetcode.cn/problems/4sum/)

题意：给定一个包含 n 个整数的数组 nums 和一个目标值 target，判断 nums 中是否存在四个元素 a，b，c 和 d ，使得 a + b + c + d 的值与 target 相等？找出所有满足条件且不重复的四元组。

**注意**：

答案中不可以包含重复的四元组。

示例： 给定数组 nums = [1, 0, -1, 0, -2, 2]，和 target = 0。 满足要求的四元组集合为： [ [-1, 0, 0, 1], [-2, -1, 1, 2], [-2, 0, 0, 2] ]

---

###  核心思想：多套一层“壳”

* **两数之和**：固定一个，找另一个（）。
* **三数之和**：固定一个 ，里面嵌套一个两数之和的双指针（）。
* **四数之和**：固定一个 ，再固定一个 ，里面嵌套一个三数之和的逻辑（）。

简单来说，就是：**用两层 `for` 循环定住两个数，剩下的两个数依然用双指针在剩下的区间里“对撞”。**

#### **思路**

**剪枝逻辑**:不能直接就`nums[i] > nums`然后`break`。

在三数之和里，我们说 `if nums[i] > 0: break`。但在四数之和里，如果 target 是 -10，而 nums[i] 是 -5，你不能因为 -5 > -10 就直接返回。因为**两个负数相加会变得更小**（比如 $-5 + (-10) = -15$），后面依然可能凑出 -10。

> 我们需要`nums[i] > target` 且 `nums[i] >= 0`

**去重逻辑**

既然有两层 `for` 循环，那么每一层都要进行去重：

* 第一层 `i` 去重：`if i > 0 and nums[i] == nums[i-1]: continue`
* 第二层 `j` 去重：`if j > i + 1 and nums[j] == nums[j-1]: continue`



````python
class Solution(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        nums.sort()
        res = []
        n = len(nums)

        for i in range(n):
            #第一层剪枝
            if nums[i] > target and nums[i] >= 0:break #直接挑出for i in range(n)
            #第一层去重
            if i >0 and nums[i] == nums[i-1]: continue #跳出这次循环

            for j in range(i+1, n):
                #第二层剪枝
                if nums[i] + nums[j] > target and nums[i] + nums[j] >= 0:break

                #第二层去重
                if j > i+1 and nums[j] == nums[j-1]:continue

                L,R = j+1,n-1
                    
                while L < R:
                    s = nums[i] + nums[j] + nums[L] + nums[R]

                    if s == target:
                        res.append([nums[i],nums[j],nums[L],nums[R]])

                        while L < R and nums[L] == nums[L + 1]: L += 1
                        while L < R and nums[R] == nums[R - 1]: R -= 1

                        L += 1
                        R -= 1
                    elif s < target:
                        L += 1
                    else:
                        R -= 1
        return res
````

````python
#答案优化版，剪枝优化

class Solution(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        res = []
        nums.sort()
        n = len(nums)

        for k in range(n-3):
            #它把当前最小的四个数相加。如果这四个加起来都比 target 大，
            # 那后面随着数字变大，更不可能凑出来了。所以直接 break。
            if nums[k] + nums[k+1] + nums[k+2] + nums[k+3] > target:
                break
            #同理最大值剪枝，如果最大值比target小直接break
            if nums[k] + nums[n-3] + nums[n-2] + nums[n-1] < target:
                continue

            if k > 0 and nums[k] == nums[k-1]:
                continue
            
            for i in range(k+1, n):
                if i+2 < n and nums[k] + nums[i] + nums[i+1] + nums[i+2] > target:
                    break
                if i < n-2 and nums[k] + nums[i] + nums[n-2] + nums[n-1] < target:
                    continue
                
                if i > k+1 and nums[i] == nums[i-1]:
                    continue
                
                l, r = i+1, n-1
                while l < r:
                    total = nums[k] + nums[i] + nums[l] + nums[r]
                    if total < target:
                        l += 1
                    elif total > target:
                        r -= 1
                    else:
                        res.append([nums[k], nums[i], nums[l], nums[r]])

                        while l < r and nums[l] == nums[l+1]:
                            l += 1
                        while l < r and nums[r] == nums[r-1]:
                            r -= 1
                        l += 1
                        r -= 1
                
        return res
````

---

### 2. 这里的“陷阱”：Target 不再是 0

在“三数之和”里，我们的目标是 `0`。但在“四数之和”里，题目要求的是 `target`。这就带来了两个需要注意的地方：

#### **A. 剪枝逻辑要变通**

在三数之和里，我们说 `if nums[i] > 0: break`。
但在四数之和里，如果 `target` 是 `-10`，而 `nums[i]` 是 `-5`，你不能因为 `-5 > -10` 就直接返回。因为**两个负数相加会变得更小**（比如 ），后面依然可能凑出 `-10`。

> **正确的剪枝**：只有当 `nums[i] > target` 且 `nums[i] >= 0` 时，才能放心断定后面没戏了。

#### **B. 去重逻辑要翻倍**

既然有两层 `for` 循环，那么每一层都要进行去重：

* 第一层 `i` 去重：`if i > 0 and nums[i] == nums[i-1]: continue`
* 第二层 `j` 去重：`if j > i + 1 and nums[j] == nums[j-1]: continue`
