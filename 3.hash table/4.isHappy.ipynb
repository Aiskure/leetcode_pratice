# 202.快乐数

编写一个算法来判断一个数 n 是不是快乐数。

「快乐数」定义为：对于一个正整数，每一次将该数替换为它每个位置上的数字的平方和，然后重复这个过程直到这个数变为 1，也可能是 无限循环 但始终变不到 1。如果 可以变为  1，那么这个数就是快乐数。

如果 n 是快乐数就返回 True ；不是，则返回 False 。

示例：

输入：19

输出：true

解释：

$1^2 + 9^2 = 82$

$8^2 + 2^2 = 68$

$6^2 + 8^2 = 100$

$1^2 + 0^2 + 0^2 = 1$



**思路**

用`divmod(a,b)`来获得每个位上的数字

```python

while n>0:
    n,digit = divmod(n,10) #n变成商，r是余数(最后一位)
    total += digit*digit

```

我们需要用`set`来判断当前位是否出现过。判断是否进入循环



````python
class Solution(object):
    def isHappy(self, n):
        """
        :type n: int
        :rtype: bool
        """
        record = set() #创建一个set，把所有出现过的结果加入
        while n != 1:
            n = self.getsum(n)
            
            if n in record:
                return False
            
            record.add(n)
        return True

    def getsum(self,n):
        total = 0
        while n >0:
            n,r = divmod(n,10)
            total += r * r

        return total

class Solution(object):
    def isHappy(self, n):
        """
        :type n: int
        :rtype: bool
        """
        dic = {n:1}
        sum = 0
        while True:
            while n>=10:
                n,r = divmod(n,10)
                sum += r**2
            sum += n**2
            if sum == 1:
                return True
            if sum in dic:
                return False
            else:
                dic[sum] = dic.get(sum,0)+1
            n = sum
            sum = 0
````
