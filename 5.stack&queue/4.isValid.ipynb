# 20.有效的括号

[力扣题目链接](https://leetcode.cn/problems/valid-parentheses/description/)

给定一个只包括 `'('`，`')'`，`'{'`，`'}'`，`'['`，`']'` 的字符串，判断字符串是否有效。

有效字符串需满足：

1. 左括号必须用相同类型的右括号闭合。
2. 左括号必须以正确的顺序闭合。
3. 注意空字符串可被认为是有效字符串。

示例 1:

- 输入: `"()"`
- 输出: true

示例 2:

- 输入: `"()[]{}"`
- 输出: true

示例 3:

- 输入:` "(]"`
- 输出: false

示例 4:

- 输入: `"([)]"`
- 输出: false

示例 5:

- 输入: `"{[]}"`
- 输出: true

---



#### 1. 核心思想：**后进先出 (LIFO)**

括号匹配的本质是：**每一个右括号，都要与离它最近的左括号进行抵消。** 这种“最近”且“对称”的结构，与栈的特性完美契合。

#### 2. 解题步骤：

1. **建立映射**：使用字典 `mapping` 存储右括号（Key）与左括号（Value）的对应关系，方便快速查找。
2. **遍历字符串**：
* **遇到左括号**：说明这是一个“待匹配”的任务，将其压入栈（`push`）。
* **遇到右括号**：
* **判空**：如果此时栈为空，说明右括号多余，直接失败。
* **匹配**：弹出栈顶元素（最近的左括号），检查它是否与当前右括号配对。若不配对，直接失败。




3. **收尾检查**：遍历结束后，检查栈是否为空。若不为空，说明左括号多余，返回失败。

#### 3. 失败的三种场景：

* **右括号多余**：栈已空，却来了右括号。
* **左右不匹配**：弹出的栈顶左括号与当前右括号对不上。
* **左括号多余**：遍历结束，栈内仍有残留。

```python
class Solution(object):
    def isValid(self, s):
        mapping = {')': '(', ']': '[', '}': '{'}
        stack = []

        for char in s:
            if char in mapping.values(): # 处理左括号
                stack.append(char)
            else: # 处理右括号
                # 1. 检查栈是否为空（右括号多了）
                # 2. 检查栈顶和当前括号是否匹配（左右不匹配）
                #如果右括号进完栈就空了，说明不匹配返回False
                #如果右括号和左括号无法匹配返回False
                #stack.pop()同时完成出栈
                if not stack or stack.pop() != mapping[char]:
                    return False
        
        # 最后一步：检查栈里是否还有剩下的左括号
        return not stack


s = '{()}'
solution = Solution()
solution.isValid(s)
```

```python
# 方法一，仅使用栈，更省空间
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        
        for item in s:
            if item == '(':
                stack.append(')')
            elif item == '[':
                stack.append(']')
            elif item == '{':
                stack.append('}')
            elif not stack or stack[-1] != item:
                return False
            else:
                stack.pop()
        
        return True if not stack else False
```
