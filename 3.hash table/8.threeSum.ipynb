# 15.三数之和

[力扣题目链接](https://leetcode.cn/problems/3sum/)

给你一个包含 n 个整数的数组 nums，判断 nums 中是否存在三个元素 a，b，c ，使得 a + b + c = 0 ？请你找出所有满足条件且不重复的三元组。

**注意**： 答案中不可以包含重复的三元组。

示例：

给定数组 nums = [-1, 0, 1, 2, -1, -4]，

满足要求的三元组集合为： [ [-1, 0, 1], [-1, -1, 2] ]

**思路**

用哈希表很难做，

一、排序加双指针

1. 排序，从小到大排好

2. 定住`i`寻找两个`L`和`R`，L为左指针，R为右指针。在nums[i]后面的区间寻找两个数，让他们的和等于-nums[i]，当nums[i] > 0 同样停止寻找。

3. 双指针对碰:计算 sum = nums[i] + nums[L] +nums[R]
    - 如果sum == 0 ：拿到一个结果，L和R移动一格
    - 如果sum <  0 ：说明太小，L往右边走一步
    - 如果sum >  0 ：说明太大，R往左走一步
    
4. 去重
    因为题目要求不能有重复的三元组，我们需要在三个地方跳过相同的数字:
    - 外层`i`的去重:如果`nums[i] == nums[i-1]`,直接`continue`
    - 内层`L`的去重:找到一个结果后，如果`nums[L] == nums[L+1]`，`L`继续往右移动
    - 内层`R`的去重:找到一个结果后，如果`nums[R] == nums[R-1]`，`R`继续往左移动

```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        nums.sort()
        res = []
        
        for i in range(len(nums)):
            #剪枝优化
            #大于0可以停止
            if nums[i] > 0:
                return res
            
            if i > 0 and nums[i] == nums[i-1]:
                continue

            L = i+1
            R = len(nums) - 1

            while L < R:
                s = nums[i] + nums[L] + nums[R]

                if s < 0:
                    L+=1

                elif s>0:
                    R-=1
                else:
                    res.append([nums[i],nums[L],nums[R]])
                    #L的去重
                    while L < R and nums[L] == nums[L + 1]:
                        L +=1

                    while L < R and nums[R] == nums[R - 1]:
                        R -=1
                    
                    #找到结果后，双指针同时想中间缩进,如果固定R不动，后两个数的和一定增大，nums[i]也增大，结果一定大于0

                    L += 1
                    R -= 1
        return res


        
```
