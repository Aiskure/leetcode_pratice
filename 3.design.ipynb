# 707.
[](https://leetcode.cn/problems/design-linked-list/)





- `get(index)` index -1

- `addAtHead(val)` val 

- `addAtTail(val)` val 

- `addAtIndex(index,val)` index  val   index  index index0

- `deleteAtIndex(index)` index  index 




```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class MyLinkedList(object):

    def __init__(self):
        self.dummy = ListNode(0)
        self.size  = 0
        
    def get(self, index):
        """
        :type index: int
        :rtype: int
        """
        if index < 0 or index >=self.size:
            return -1
        
        curr = self.dummy.next #head
        for _ in range(index): #O(n)
            curr = curr.next
        
        return curr.val

    def addAtHead(self, val):
        """
        :type val: int
        :rtype: None
        """
        self.addAtIndex(0, val)

    def addAtTail(self, val):
        """
        :type val: int
        :rtype: None
        """
        self.addAtIndex(self.size,val)

    def addAtIndex(self, index, val):
        """
        :type index: int
        :type val: int
        :rtype: None
        """
        if index < 0 :
            index = 0 #
        
        if index > self.size: #index == size
            return #

        

        #indexindex - 1

        curr = self.dummy #head - 1

        for _ in range(index):
            curr = curr.next

        newnode = ListNode(val)
        newnode.next = curr.next
        curr.next = newnode
        
        self.size += 1 #


    def deleteAtIndex(self, index):
        """
        :type index: int
        :rtype: None
        """
        if index < 0 or index >= self.size: #index0size
            return

        curr = self.dummy
        for _ in range(index): #index-1
            curr = curr.next

        curr.next = curr.next.next
        
        self.size -= 1


        


# Your MyLinkedList object will be instantiated and called as such:
# obj = MyLinkedList()
# param_1 = obj.get(index)
# obj.addAtHead(val)
# obj.addAtTail(val)
# obj.addAtIndex(index,val)
# obj.deleteAtIndex(index)
```

```python
def print_linked_list(obj):
    """"""
    nodes = []
    curr = obj.dummy.next  # 
    while curr:
        nodes.append(str(curr.val))
        curr = curr.next
    print(f" (size={obj.size}): " + " -> ".join(nodes) if nodes else "")
```

```python
# 1. 
obj = MyLinkedList()
print("---  ---")

# 2. 
obj.addAtHead(1)      # : 1
obj.addAtHead(2)      # : 2 -> 1
print(" 1  2 ")
print_linked_list(obj)

# 3. 
obj.addAtTail(3)      # : 2 -> 1 -> 3
print(" 3 ")
print_linked_list(obj)

# 4. 
obj.addAtIndex(1, 4)  #  1  4: 2 -> 4 -> 1 -> 3
print(" 1  4 ")
print_linked_list(obj)

# 5. 
val = obj.get(1)
print(f" 1 : {val} (: 4)")

# 6. 
obj.deleteAtIndex(1)  #  1: 2 -> 1 -> 3
print(" 1 ")
print_linked_list(obj)

# 7. 
obj.deleteAtIndex(5)
print(" 5 ")
print_linked_list(obj)
```
