使用栈实现队列的下列操作：

`push(x)` -- 将一个元素放入队列的尾部。<br/>
`pop()` -- 从队列首部移除元素。<br/>
`peek()` -- 返回队列首部的元素。<br/>
`empty()` -- 返回队列是否为空。<br/>

```python
class MyQueue(object):

    def __init__(self):
        self.in_stack = [] #入列
        self.out_stack = []#出列

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        self.in_stack.append(x)#入栈直接导入

    def pop(self):
        """
        :rtype: int
        """

        if not out_stack : #out_stack空时
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())#我们把in_stack的内容全部倒过来


        return self.out_stack.pop()

    def peek(self):
        """
        :rtype: int
        """
        #返回
        if not out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack[-1]

    def empty(self):
        """
        :rtype: bool
        """
        return not self.in_stack and not self.out_stack
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```
